```python
#create 2 definitons one to define leap year and other to days in a month
month_days = [0,31,28,31,30,31,30,31,31,30,31,30,31]
def leap_year(year):
    return year%4==0 and (year % 100 != 0 or year % 400 == 0)

print(leap_year(2000))

def days_mon(year,month):
    if not 1 <= month <=12:
        return 'invalid month'
    if month == 2 and leap_year(year):
        return 29
    
    return month_days[month]
print(days_mon(2025,2))


```

```python
for i in customers:
    
    print(i["name"],sum(i["orders"]))
    if sum(i["orders"]) > 250:
        print("High value customer")
```

```python
for i in customers:
    at=sum(i["orders"])
    label = "HIGH VALUE CUSTOMER" if at > 250 else ""
    print(i["name"],at,label)
    
```

```python
def sum_orders(*args):
    return sum(args)
for i in customers:
    at=sum_orders(*i["orders"])
    label = "HIGH VALUE CUSTOMER" if at > 250 else ""
    print(i["name"],at,label)
```
