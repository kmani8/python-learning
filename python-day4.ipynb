```python
def load_data(file_path):
    df=pd.read_csv("/Users/mani/orders.csv")
    return df
def clean_data(df):
    df["amount"] = pd.to_numeric(df["amount"],errors="coerce")
    df=df.dropna()
    return df
def calculate_metrics(df):
    total = df["amount"].sum()
    grouped = df.groupby('customer')["amount"].sum()
    return total,grouped
def save_summary(file_path,total,grouped):
    with open(file_path,"w") as f:
        f.write(f"Total Revenue:{total}\n")
        f.write(f"Revenue per Customer:\n")
        for customer, revenue in grouped.items():
            f.write(f"{customer}: {revenue}\n")
    
file_path="/Users/mani/orders.csv"
df=load_data(file_path)
df=clean_data(df)
calculate_metrics(df)
total,grouped=calculate_metrics(df)
save_summary("customer_summary.txt",total,grouped)
```
