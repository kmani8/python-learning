```python
def read_orders(file_path):
    values=[]
    with open(file_path,"r") as f:
        for i in f:
            values.append(i.strip())
    return values

def clean_orders(values):
    cleaned_values=[]
    for i in values:
        try:
            cleaned_values.append(int(i))
        except:
            continue
    return cleaned_values


def calculate_total(orders):
    return sum([i for i in orders])

def write_summary(file_path,total_orders,revenue):
    with open(file_path,"w") as f:
        f.write(f"Total Orders:{total_orders}\n")
        f.write(f"Total Revenue:{revenue}\n")

data = read_orders("processed_orders.txt")
cleaned=clean_orders(data)
total_orders=len(cleaned)
revenue=calculate_total(cleaned)

write_summary("summary.txt",total_orders,revenue)
```
